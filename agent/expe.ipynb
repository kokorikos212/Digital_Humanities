```python
from pydantic import BaseModel, Field

class GetWeatherInput(BaseModel):
    """Get the current weather in a given location"""
    location: str = Field(..., description="The city and state, e.g. San Francisco, CA")
```

```python
print(GetWeatherInput.model_json_schema())
```

```python
import spacy
from spacy import displacy

nlp = spacy.load("en_core_web_sm")
doc = nlp("This sophisticated graph illustrates complex linguistic patterns beautifully.")

options = {
    "compact": True,           # Square arrows for a clean, structural look
    "font": "Courier New",      # Technical, typewriter aesthetic
    "distance": 140,           # Wide spacing to allow the structure to breathe
    "offset_x": 50,
    "color": "#222222",        # Off-black for a softer ink look
    "bg": "#ffffff"
}

displacy.serve(doc, style="dep", options=options)
```

```python
import spacy
from spacy import displacy

nlp = spacy.load("en_core_web_sm")
doc = nlp("This sophisticated graph illustrates complex linguistic patterns beautifully.")

options = {
    "compact": False,      # Rounded arcs
    "distance": 110,       # Standard height for better density
    "font": "Verdana",     # Clean, wide sans-serif for legibility
    "color": "#F0F0F0",    # Light grey arcs and text
    "bg": "#2C3E50",       # Deep slate blue/grey background
    "offset_x": 50
}
displacy.serve(doc, style="dep", options=options)
```

```python
doc = nlp("Apple is looking at buying U.K. startup for $1 billion")

for ent in doc.ents:
    print(ent.text, ent.start_char, ent.end_char, ent.label_)
```
